```python
import google.generativeai as genai
from google.colab import userdata

# Load API key from Colab Secrets (NOT hardcoded)
API_KEY = userdata.get('GEMINI_API_KEY')
genai.configure(api_key=API_KEY)

MODEL_NAME = "gemini-2.5-flash"
```

```python
def generate_blog(topic, tone, word_count):
    """Generate a blog post on a given topic, tone, and word count."""
    system_instruction = (
        "You are an experienced blog writer who crafts engaging, well-structured "
        "blog posts. Always include a title, an intro hook, 2-3 body sections "
        "with subheadings, and a conclusion."
    )

    model = genai.GenerativeModel(
        model_name=MODEL_NAME,
        system_instruction=system_instruction
    )

    prompt = (
        f"Write a blog post on the topic: \"{topic}\".\n"
        f"Tone: {tone}.\n"
        f"Target word count: approximately {word_count} words.\n\n"
        "Structure the output with:\n"
        "- A compelling title\n"
        "- An intro hook\n"
        "- 2-3 body sections with subheadings\n"
        "- A conclusion"
    )

    response = model.generate_content(
        prompt,
        generation_config=genai.types.GenerationConfig(temperature=0.7)
    )

    blog_text = response.text

    with open("blog_output.txt", "w") as f:
        f.write(blog_text)

    return blog_text
```

```python
def main():
    while True:
        print("\n===== BLOG GENERATOR MENU =====")
        print("1. Generate Blog")
        print("2. Exit")
        choice = input("Choose an option (1-2): ").strip()

        if choice == "1":
            topic = input("Enter blog topic: ")
            tone = input("Enter tone (formal / casual / Gen Z / professional): ")
            word_count = input("Enter target word count: ")

            print("\nGenerating blog...\n")
            blog = generate_blog(topic, tone, word_count)

            print("===== BLOG GENERATOR =====")
            print(f'Topic: "{topic}"')
            print(f"Tone: {tone}")
            print(f"Word count: {word_count}\n")
            print(blog)
            print("\n(Saved to blog_output.txt)")

        elif choice == "2":
            print("Goodbye!")
            break

        else:
            print("Invalid choice, try again.")
```

```python
main()
```
