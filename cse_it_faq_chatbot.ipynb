```python
import google.generativeai as genai
from google.colab import userdata

# Load API key from Colab Secrets (NOT hardcoded)
API_KEY = userdata.get('GEMINI_API_KEY')
genai.configure(api_key=API_KEY)

MODEL_NAME = "gemini-2.5-flash"
```

```python
knowledge_base = """
1. Q: What programs does the CSE/IT department offer?
   A: The department offers a B.Tech in Computer Science and Engineering, with specialization tracks including Data Science, AI/ML, and Cloud Computing.

2. Q: How long is the B.Tech program?
   A: The B.Tech program is 4 years, split into 8 semesters.

3. Q: What is the minimum CGPA required to stay in good academic standing?
   A: Students must maintain a minimum CGPA of 5.0 each semester to avoid academic probation.

4. Q: What core subjects are covered in the first two years?
   A: Core subjects include Data Structures, Algorithms, Operating Systems, DBMS, Computer Networks, OOP, and Discrete Mathematics.

5. Q: Does the department offer electives in AI/ML?
   A: Yes, electives include Machine Learning, Deep Learning, NLP, Computer Vision, and Reinforcement Learning, typically offered from the 5th semester onward.

6. Q: Is there a mandatory internship requirement?
   A: Yes, students must complete at least one internship of a minimum of 6 weeks before their final year, which is evaluated for credit.

7. Q: What labs does the department have?
   A: The department has a Programming Lab, Networks Lab, AI/ML Lab, Cloud Computing Lab, and a Project Lab for capstone work.

8. Q: How are final year projects evaluated?
   A: Final year projects are evaluated through a proposal review, mid-term review, and a final demo with a report, judged by a faculty panel.

9. Q: What programming languages are taught?
   A: Python, Java, and C/C++ are taught as part of the core curriculum, with additional exposure to SQL and JavaScript in relevant courses.

10. Q: Does the department support student clubs or communities?
    A: Yes, the department supports a Coding Club, an AI/ML interest group, and a Competitive Programming club.

11. Q: What is the attendance requirement to sit for exams?
    A: Students need a minimum of 75% attendance per course to be eligible to sit for end-semester exams.

12. Q: How can a student request a grade recheck?
    A: Students can submit a written request to the department office within 7 days of grades being published, along with the applicable recheck fee.

13. Q: Are there opportunities for research under faculty?
    A: Yes, students can apply for faculty-guided research projects starting from their 3rd year through the department's research portal.

14. Q: What placement support does the department provide?
    A: The department conducts resume workshops, mock interviews, and coordinates with the placement cell for on-campus drives and referrals.

15. Q: Who do I contact for academic issues or grievances?
    A: Academic grievances should be directed to the department's Academic Coordinator via email or during posted office hours.

16. Q: Can students switch specialization tracks mid-program?
    A: Track switches are allowed only at the end of the 4th semester, subject to seat availability and minimum CGPA criteria.
"""
```

```python
system_instruction = f"""
You are the official FAQ assistant for the CSE/IT Department.

IDENTITY:
- You represent the CSE/IT department and help students, parents, and visitors
  with questions about academics, labs, electives, projects, and department policies.

RULES:
- Answer ONLY using the knowledge base provided below. Do not invent facts.
- If a question cannot be answered from the knowledge base, politely say you
  don't have that information and suggest the student contact the department office.
- Keep answers concise, clear, and friendly.
- Maintain context across the conversation (multi-turn memory).
- Do not answer questions unrelated to the CSE/IT department (e.g. general
  knowledge, other departments, personal opinions) — politely decline and
  redirect to department-related topics.

KNOWLEDGE BASE:
{knowledge_base}
"""
```

```python
def main():
    model = genai.GenerativeModel(
        model_name=MODEL_NAME,
        system_instruction=system_instruction,
        generation_config=genai.types.GenerationConfig(temperature=0.2)
    )

    chat = model.start_chat(history=[])

    print("===== CSE/IT DEPARTMENT FAQ CHATBOT =====")
    print("Ask me anything about the CSE/IT department. Type 'quit' or 'exit' to leave.\n")

    while True:
        user_input = input("You: ").strip()

        if user_input.lower() in ["quit", "exit"]:
            print("Bot: Thanks for chatting! Goodbye.")
            break

        if not user_input:
            continue

        response = chat.send_message(user_input)
        print(f"Bot: {response.text}\n")
```

```python
main()
```
